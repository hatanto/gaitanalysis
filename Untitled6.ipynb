```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv(GRF_Wallking_Speed_CVS)  # Ganti 'nama_file.csv' dengan nama file data Anda

# Menampilkan persebaran data untuk setiap pasien pada sesi yang sama
fig, ax = plt.subplots(figsize=(20, 6))

# Iterasi melalui setiap pasien dan sesi yang sama
for subject_session, group in data.groupby(['SUBJECT_ID', 'SESSION_ID']):
    subject_id, session_id = subject_session
    plt.scatter(group['TRIAL_ID'], group['WALKING_SPEED'], label=f'Pasien {subject_id}, Sesi {session_id}')

# Menambahkan label dan legenda
plt.xlabel('TRIAL_ID')
plt.ylabel('WALKING_SPEED')

# Menampilkan plot
plt.show()
```
